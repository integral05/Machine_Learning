In this exercise, we are using Pandas and Matplotlib to visualize Company Sales Data.



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("company_sales_data.csv")
df
```

Exercise 1: Read Total profit of all months and show it using a line plot

```python
plt.title("Total Profit of All Months")
plt.xlabel("Month Number")
plt.ylabel("Profit")
plt.xticks(df['month_number'])
plt.plot(df['month_number'],df['total_profit'])
```

Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: –


Line Style dotted and Line-color should be. red
Show legend at the lower right location.
X label name = Month .Number
Y label name = Sold units. number
Add a circle marker.
Line marker colo.r as read

```python
plt.title("Profit in Every Month")
plt.xlabel('Month Number')
plt.ylabel('Profit')
plt.plot(df['month_number'],df['total_profit'],'ro--',linewidth=3,label= 'Company Total Profit')  
##Here color = red, marker = o, linestyle = 'dashed'
plt.legend(loc='lower right')
```

```python
plt.title('Sales Data')
plt.xlabel('Number of Months')
plt.ylabel('Number of Sales')
plt.plot(df['month_number'],df['facecream'],'o-',label = 'Facecream')
plt.plot(df['month_number'],df['facewash'],'o-',label = 'Facewash')
plt.plot(df['month_number'],df['toothpaste'],'o-',label = 'Toothpaste')
plt.plot(df['month_number'],df['bathingsoap'],'o-',label = 'Bathing Soap')
plt.plot(df['month_number'],df['shampoo'],'o-',label = 'Shampoo')
plt.plot(df['month_number'],df['moisturizer'],'o-',label = 'Moisturizer')
plt.xticks(df['month_number'])
plt.legend()
```
